```python
import time
import numpy as np
from bokeh.plotting import Session, cursession, output_server, show, figure

from data.datagenerator import normal_list
```

```python
session = Session(load_from_config=False, root_url = "http://127.0.0.1:5006/")
output_server("candlestick", url = "http://127.0.0.1:5006/")
```

```python
class Histogram(object):
    
    def __init__(self, name = "histogram", bins = 10):
        self.bins = bins
        self.data = []
        self.plot = figure(plot_width = 900,
                           plot_height = 300,
                           name = name,
                           title = name)
        self.plot.quad(top = [0 for i in range(self.bins)], 
                       bottom = [0 for i in range(self.bins)],
                       left = [0 for i in range(self.bins)], 
                       right = [0 for i in range(self.bins)],
                       line_color="black",
                       fill_alpha=0.2,
                       name = "quads"
                      ) 
        self.ds = self.plot.select({"name":"quads"})[0].data_source
        
    def get_plot(self):
        return self.plot
    
    def update(self, data):
        self.data += data
        values, edges = np.histogram(self.data, self.bins)
        self.ds.data["top"] = list(values)
        self.ds.data["left"] = list(edges[:-1])
        self.ds.data["right"] = list(edges[1:])
        print(self.ds.data["top"])
        print(self.ds.data["left"])
        print(self.ds.data["right"])
        print(self.ds.data["bottom"])
        
        return [self.ds]
```

```python
bins = 20
h = Histogram("hist", bins)
show(h.get_plot())
cursession().store_objects(*h.update([i for i in range(bins)])) #inizializza a 1 tutti i bin in un sol colpo
for i in range(bins):
    cursession().store_objects(*h.update([i])) #incrementa di 1 il valore in ciascun bin, un bin alla volta
    time.sleep(0.5)
for i in range (100):
    cursession().store_objects(*h.update(normal_list(10, 10, 5, 0))) #aggiunge 10 valori casuali alla volta
    time.sleep(0.5)  
```
